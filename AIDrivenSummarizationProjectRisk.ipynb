```python
import subprocess, threading, os
def stream_process(cmd, logfile=None):
    if logfile is None: logfile = os.path.join(os.getcwd(), 'meeting_mcp.log')
    proc = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True, bufsize=1)
    def reader():
        try:
            with open(logfile, 'a', encoding='utf-8') as f:
                for line in proc.stdout:
                    print(line, end='')
                    f.write(line)
                    f.flush()
        except Exception as e:
            print('Error streaming process output:', e)
    t = threading.Thread(target=reader, daemon=True)
    t.start()
    return proc, t

print('stream_process helper ready — use it to start services for live notebook logging')
```

```python
import socket, sys, os
def check_port(port, host='127.0.0.1'):
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    try: s.settimeout(1.0); res = s.connect_ex((host, port)); return res == 0
    finally: s.close()
ports = {'streamlit':8501, 'uvicorn':8000}
for name, p in ports.items(): print(f'Port {p} for {name}:', 'in use' if check_port(p) else 'available')
if sys.platform.startswith('win'):
    print('On Windows use Task Manager or `taskkill /PID <pid> /F` to kill processes.')
else:
    print('On Linux/macOS use `lsof -i :<port>` or `kill <pid>` to stop the process.')
```

```python
# 9) Start Streamlit and Uvicorn under stream_process (run after steps above)
import os
from pyngrok import ngrok

# Start ngrok tunnel
ngrok_tunnel = ngrok.connect(8501)
print(f'ngrok tunnel URL: {ngrok_tunnel.public_url}')

streamlit_cmd = ['streamlit', 'run', os.path.join('meeting_mcp','ui','streamlit_agent_client.py'), '--server.port', '8501', '--server.headless', 'true']
uvicorn_cmd = ['uvicorn', 'meeting_mcp.server.mcp_api:app', '--host', '127.0.0.1', '--port', '8000']

print('Starting Streamlit...')
streamlit_proc, streamlit_thread = stream_process(streamlit_cmd, logfile=os.path.join(os.getcwd(),'meeting_mcp.log'))
print('Streamlit PID:', getattr(streamlit_proc, 'pid', None))
print('Starting Uvicorn...')
uvicorn_proc, uvicorn_thread = stream_process(uvicorn_cmd, logfile=os.path.join(os.getcwd(),'meeting_mcp.log'))
print('Uvicorn PID:', getattr(uvicorn_proc, 'pid', None))

__PROJECT_SERVICES__ = {'streamlit': streamlit_proc, 'uvicorn': uvicorn_proc}
print('Services started. To stop: call stop_services() or terminate specific procs stored in __PROJECT_SERVICES__')
```

```python
def stop_services():
    procs = globals().get('__PROJECT_SERVICES__', {})
    if not procs:
        print("No services found to stop.")
        return
    for name, p in procs.items():
        try:
            print(f"Terminating {name} (PID: {getattr(p, 'pid', None)})")
            p.terminate()  # Attempt to terminate the process
            p.wait(timeout=5)  # Wait for the process to terminate
            print(f"{name} terminated successfully.")
        except Exception as e:
            print(f"Error terminating {name}: {e}")
            try:
                p.kill()  # Force kill if terminate fails
                print(f"{name} killed forcefully.")
            except Exception as kill_error:
                print(f"Failed to kill {name}: {kill_error}")
```

```python
# 10) Tail the combined log file (cross-platform, fixed)
import os
def tail_file(path, n=50):
    try:
        with open(path, 'rb') as f:
            f.seek(0, os.SEEK_END)
            filesize = f.tell()
            blocksize = 1024
            data = b''
            while filesize > 0 and data.count(b'\n') <= n:
                readsize = min(blocksize, filesize)
                f.seek(filesize - readsize)
                data = f.read(readsize) + data
                filesize -= readsize
            lines = data.splitlines()[-n:]
            for line in lines:
                try: print(line.decode('utf-8'))
                except: print(line)
    except FileNotFoundError: print('Log file not found:', path)
tail_file(os.path.join(os.getcwd(), 'meeting_mcp.log'), n=100)
```

```python
# Colab connectivity test for MCP server
import os, requests, json, pprint

MCP_SERVER_URL = "http://127.0.0.1:8000"  # change if server hosted elsewhere
MCP_API_KEY = os.environ.get("MCP_API_KEY")  # or set a string here

def headers():
    h = {"Content-Type": "application/json"}
    if MCP_API_KEY:
        h["Authorization"] = f"Bearer {MCP_API_KEY}"
    return h

pp = pprint.PrettyPrinter(indent=2)

# Health
try:
    r = requests.get(f"{MCP_SERVER_URL}/health", headers=headers(), timeout=5)
    print("Health:", r.status_code); pp.pprint(r.json())
except Exception as e:
    print("Health check failed:", e)

# Create session
session_id = None
try:
    r = requests.post(f"{MCP_SERVER_URL}/session/create", json={"agent_id": "colab-test"}, headers=headers(), timeout=10)
    print("Create session:", r.status_code); pp.pprint(r.json())
    session_id = r.json().get("session_id")
except Exception as e:
    print("Create session failed:", e)

# Orchestrate (simple ping)
try:
    payload = {"prompt": "ping", "params": {}}
    if session_id:
        payload["session_id"] = session_id
    r = requests.post(f"{MCP_SERVER_URL}/mcp/orchestrate", json=payload, headers=headers(), timeout=30)
    print("Orchestrate:", r.status_code); pp.pprint(r.json())
except Exception as e:
    print("Orchestrate failed:", e)

# End session
if session_id:
    try:
        r = requests.post(f"{MCP_SERVER_URL}/session/{session_id}/end", headers=headers(), timeout=5)
        print("End session:", r.status_code); pp.pprint(r.json())
    except Exception as e:
        print("End session failed:", e)
```
